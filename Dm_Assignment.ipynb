```python
import numpy as np 
import pandas as pd # pandas is so commonly used, it's shortened to pd import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
```

```python
phone = pd.read_csv("/content/drive/MyDrive/assg/train.csv") 
phone.head(10)
```

```python
phone.isna().sum()
```

```python
from sklearn.impute import SimpleImputer 
from sklearn.compose import ColumnTransformer
#Fill categorical values with 'missing' & numerical values with mean
num_imputer=SimpleImputer(strategy="mean")
#Define columns
num_features = ["battery_power","clock_speed","four_g","int_memory","n_cores","pc","ram","touch_screen","wifi","price_range"]
# Create an imputer (Something that fills missing data)
imputer = ColumnTransformer([ ("num_imputer",num_imputer,num_features)
])
# Transformed the Data
filled_X = imputer.fit_transform(phone)
filled_X
```

```python
phone = pd.DataFrame(filled_X,columns=["battery_power","clock_speed","four_g","int_memory","n_cores","pc","ram","touch_screen","wifi","price_range"])
phone.head(10)
```

```python
phone.isna().sum()
```

```python

X=phone.drop("price_range",axis=1)
y=phone["price_range"]
```

```python
np.random.seed(69)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

model = RandomForestRegressor() 
model.fit(X_train,y_train) 
model.score(X_test,y_test)
```

```python
 y_preds = model.predict(X_test)
```

```python
output = pd.DataFrame({"ram":X_test.ram,'price_range?':y_preds}) 
output
```
